# **Data Cleaning**

```python
df["Production (Tons)"] = df["Production (Tons)"].str.replace(".","")
df["Production per Person (Kg)"] = df["Production per Person (Kg)"].str.replace(".","").str.replace(",",".")
df["Acreage (Hectare)"] = df["Acreage (Hectare)"].str.replace(".","")
df["Yield (Kg / Hectare)"]= df["Yield (Kg / Hectare)"].str.replace(".","").str.replace(",",".")
```

```python
df.head()
```

```python
df = df.drop("Unnamed: 0", axis = 1)
```

```python
df.head()
```

```python
df.rename(columns = {"Production (Tons)" : "Production(Tons)"}, inplace = True)
df.rename(columns = {"Production per Person (Kg)" : "Production_per_person(Kg)"}, inplace = True)
df.rename(columns = {"Acreage (Hectare)" : "Acreage(Hectare)"}, inplace = True)
df.rename(columns = {"Yield (Kg / Hectare)" : "Yield(Kg/Hectare)"}, inplace = True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
df["Production(Tons)"] = df["Production(Tons)"].astype(float)
df["Production_per_person(Kg)"] = df["Production_per_person(Kg)"].astype(float)
df["Acreage(Hectare)"] = df["Acreage(Hectare)"].astype(float)
df["Yield(Kg/Hectare)"] = df["Yield(Kg/Hectare)"].astype(float)
```

```python
df.dtypes
```

# **Analysis of continent and total sugarcane production**

```python
df.groupby('Continent').sum()['Production(Tons)'].sort_values().plot(kind = 'bar', figsize = (6,6), ylabel = 'Total Production(Tons)')
plt.title("Continents with least to most production", fontdict = {"fontsize" : 16})
```

South America is the continent which produces maximum sugarcane.

# **Production distribution by continent**

```python
df.groupby('Continent').sum()['Production(Tons)'].plot(kind = 'pie', figsize = (6,6), autopct = "%0.2f%%")
```

South America produces 45.86% of the total sugarcane produced.

# **How many countries produce sugarcane from each continent?**

```python
df.groupby('Continent').size().sort_values(ascending = False)
```

```python
sns.set(rc = {'figure.figsize' : (10,5)})
sns.countplot(x = 'Continent', data = df, order = df['Continent'].value_counts(ascending = True).index)
```

Africa is having the most number of countries which produces sugarcane.

# **Analysis of country and total sugarcane production**

```python
# Displaying all countries along with their total sugarcane production from maximum to minimum

df.groupby('Country').sum()['Production(Tons)'].sort_values(ascending = False)
```

```python
# Top 5 country with maximum sugarcane production

df.groupby('Country').sum()['Production(Tons)'].sort_values(ascending = False).head()
```

```python
df.groupby('Country').sum()['Production(Tons)'].sort_values(ascending = False).head().plot(kind = "bar", ylabel = "Production(Tons)")
plt.title("Countries with most to least production", fontdict = {"fontsize" : 16})
```

Brazil produces the maximum quantity of sugarcane.

# **Analysis of Continent and Yield(Kg/Heactare)**

```python
df.groupby('Continent').sum()['Yield(Kg/Hectare)'].sort_values().plot(kind = 'bar', ylabel = 'Yield(Kg/Hectare)')
plt.title("Continents with least to most Yield(Kg/Hectare)", fontdict = {"fontsize" : 16})
```

Africa is having the maximum Yield(Kg/Hectare) .
